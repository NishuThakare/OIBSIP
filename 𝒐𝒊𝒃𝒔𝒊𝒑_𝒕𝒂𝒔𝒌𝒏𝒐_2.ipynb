```python
#Plotting the boxplot of the price data.
plt.figure(figsize=(10,5))
ax = sns.boxplot(y='price', data=task2_data_new).set_title('Price Distribution by neighbourhood group')
sns.set_theme(style='white')
plt.xlabel('Neighbourhood')
plt.ylabel('Price')
plt.show()
```

As expected, there are outliers present in the data. To address this, we will apply quantile-based flooring and capping. First, let's display the 10th and 90th percentiles of the price column.

```python
#finding the low quantile
q_low = task2_data_new['price'].quantile(0.10)
q_low
```

```python
# finding the high quantile
q_high = task2_data_new['price'].quantile(0.9)
q_high
```

Looks like the 10th percentile is 49USD and the 90th percentile is 269USD. Since we are dealing with Airbnb listings, it is not uncommon for a few places to be much more expensive than others, especially luxurious ones.

Given that we are going to perform calculations in our data and the mean can be skewed by outliers, let's go ahead and implement the above technique by removing the values that don't apply to the specified range – set between 49USD and 269USD.

```python
#removing the values below 10th percentile and above 90th percentile
task2_data_new = task2_data_new.drop(task2_data_new[task2_data_new['price']<q_low].index)
task2_data_new = task2_data_new.drop(task2_data_new[task2_data_new['price']>q_high].index)
```

```python
#Lets create a new boxplot and check the result.
plt.figure(figsize=(10,5))
ax = sns.boxplot(y='price', data=task2_data_new).set_title('Price Distribution')
sns.set_theme(style='white')
plt.ylabel('Price')
plt.show()
```

```python
# Now lets see the price distribution by neighbourhood gropus
plt.figure(figsize=(10,5))
ax = sns.violinplot(x="neighbourhood_group", y="price", data=task2_data_new).set_title('Price Distribution by neighbourhood groups')
plt.show()
```

The violin plot reveals that Manhattan has the highest price range, making it the most expensive area. Brooklyn follows with the second-highest rental prices, while the Bronx stands out as the most affordable option.

The neighborhoods with the highest number of listings are in Manhattan and Brooklyn, which is expected as these areas are popular among tourists. Williamsburg (Brooklyn) ranks first with 3,415 listings, while in Manhattan, Bedford-Stuyvesant has the most listings, totaling 2,311 offers.

### Key Insights from the Analysis  

- **Harlem** in **Manhattan** is the most popular neighborhood, with **Private Room** being the preferred room type.  
- **Manhattan and Brooklyn** together account for **87% of all listings** in NYC, making them the most in-demand areas.

### Conclusion  

- **Manhattan** leads in both the **number of listings and rental prices**, largely due to its **famous attractions** such as museums, theaters, and shopping districts, which sustain high rental demand throughout the year.  

- **Brooklyn** is the **second-largest rental market**, offering a substantial number of listings at **more affordable prices** compared to Manhattan.  

- **Queens, Bronx, and Staten Island** have **fewer listings**, with **Staten Island** being the least active due to its primarily **residential nature**. Many guests may choose these areas for **budget-friendly options** or to **stay closer to family and friends**.
