# Exploratory Data Analysis

```python
# Time Series Analysis: Monthly Sales Trend
df1['month'] = df1['date'].dt.to_period('M')
monthly_sales = df1.groupby('month')['sales'].sum()
plt.figure(figsize=(12,5))
monthly_sales.plot(title='Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

```python
result=df1.groupby("product_id")[["revenue","stock"]].sum()
print(result.head())
```

# Total sales group by product_id

```python
total_sales_by_product = df1.groupby('product_id')['sales'].sum()

# Display the result
print(total_sales_by_product)
```

# Total Profit Group by Product_id

```python
# Calculate the cost (stock * price)
df1['cost'] = df1['stock'] * df1['price']

# Calculate profit (revenue - cost)
df1['profit'] = df1['revenue'] - df1['cost']

# Group by 'product_id' and calculate the total profit
total_profit_by_product = df1.groupby('product_id')['profit'].sum()

# Display the result
print(total_profit_by_product)
```

# Find the product with the maximum profit

```python
max_profit_product = total_profit_by_product.idxmax()
max_profit_value = total_profit_by_product.max()
print(f"The product with the maximum profit is {max_profit_product} with a profit of {max_profit_value}.")
```

```python
# Plot histograms for sales, revenue, stock, price
df1[['sales', 'revenue', 'stock', 'price']].hist(figsize=(10, 6), bins=5)
plt.tight_layout()
plt.show()
```

```python
# Correlation matrix for numerical features
corr_matrix = df1[['sales', 'revenue', 'stock', 'price']].corr()

# Plot the correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
